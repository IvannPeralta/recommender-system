```python
import matplotlib.pyplot as plt

modelos = ["UserKNN", "SVD", "MF", "PMF"]
rmse = [1.0194, 0.9354, 0.9239, 1.0140]
mae =  [0.8038, 0.7376, 0.7507, 0.8266]

x = range(len(modelos))
width = 0.35

plt.bar([p - width/2 for p in x], rmse, width, label='RMSE')
plt.bar([p + width/2 for p in x], mae, width, label='MAE')

plt.xticks(x, modelos)
plt.ylabel("Error")
plt.title("Comparaci√≥n de RMSE y MAE por modelo")
plt.legend()
plt.show()
```

```python
precision = [0, 0, 0.0732, 0.0506]
ndcg =     [0, 0, 0.0762, 0.0492]

plt.bar([p - width/2 for p in x], precision, width, label='Precision@5')
plt.bar([p + width/2 for p in x], ndcg, width, label='NDCG@5')

plt.xticks(x, modelos)
plt.ylabel("Valor")
plt.title("Precision@5 y NDCG@5 por modelo")
plt.legend()
plt.show()
```

```python
import numpy as np

labels = ['RMSE', 'MAE', 'Precision@5', 'NDCG@5']
data = [
    [1.0194, 0.8038, 0,      0     ],  # UserKNN
    [0.9354, 0.7376, 0,      0     ],  # SVD
    [0.9239, 0.7507, 0.0732, 0.0762],  # MF
    [1.0140, 0.8266, 0.0506, 0.0492]   # PMF
]

data = np.array(data)
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
data = np.concatenate((data, data[:,[0]]), axis=1)
angles += angles[:1]

plt.figure(figsize=(6,6))
for i, model in enumerate(modelos):
    plt.polar(angles, data[i], label=model)
plt.xticks(angles[:-1], labels)
plt.title("Comparaci√≥n general de modelos")
plt.legend()
plt.show()
```
